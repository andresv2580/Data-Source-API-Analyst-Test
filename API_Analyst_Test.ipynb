```python
import requests
from requests.auth import HTTPBasicAuth

# Your GitHub token (go to GitHub Settings > Developer settings > Personal access tokens)
token = "real_token"
headers = {
    "Authorization": f"token {token}",
    "Accept": "application/vnd.github.v3+json"
}
```

```python
response = requests.get("https://api.github.com/user", headers=headers)
print(response.json())
```

```python
def search_repositories(query):
    url = "https://api.github.com/search/repositories"
    params = {"q": query}
    response = requests.get(url, headers=headers, params=params)
    return response.json()

print(search_repositories("python"))
```

```python
def get_commits(owner, repo):
    url = f"https://api.github.com/repos/{owner}/{repo}/commits"
    response = requests.get(url, headers=headers)
    return response.json()

print(get_commits("torvalds", "linux"))
```

```python
def get_contents(owner, repo, path=""):
    url = f"https://api.github.com/repos/{owner}/{repo}/contents/{path}"
    response = requests.get(url, headers=headers)
    return response.json()

print(get_contents("torvalds", "linux", "README"))
```

```python
def make_github_request(url, headers, params=None):
    try:
        response = requests.get(url, headers=headers, params=params)
        response.raise_for_status()  # Raises error for 4XX/5XX codes
        return response.json()
    except requests.exceptions.HTTPError as err:
        print(f"HTTP Error: {err}")
        return None
    except Exception as err:
        print(f"Other error: {err}")
        return None
```

```python
def get_all_commits(owner, repo):
    commits = []
    page = 1
    while True:
        url = f"https://api.github.com/repos/{owner}/{repo}/commits?page={page}"
        response = requests.get(url, headers=headers)
        data = response.json()
        if not data:
            break
        commits.extend(data)
        page += 1
    return commits
```
